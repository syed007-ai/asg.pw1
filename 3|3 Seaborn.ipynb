1) Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot

- sns.heatmap - Heatmaps are used to represent the correlation or relationship between multiple variables. They display a matrix of colors where each cell represents the strength of the relationship between two variables. Heatmaps are commonly used in exploratory data analysis and feature correlation analysis. 
- sns.histplot - Histograms are used to visualize the distribution of a single variable. They provide insights into the shape, center, and spread of the data. Histograms are particularly helpful for understanding the data's skewness or presence of outliers.
- sns.displot - Distribution plot, plots the whole dataset. It allows you to visualize the distribution of a single variable or compare the distributions of multiple variables
- sns.catplot - The catplot (Categorical Plot) is another powerful plot in Seaborn that allows you to visualize relationships between categorical variables. It provides a general interface for creating various types of categorical plots.
- sns.barplot - Bar plots are effective for comparing categorical variables or showing the distribution of a categorical variable. They are commonly used to display counts, frequencies, or proportions across different categories.
- sns.boxplot - Box plot are very useful in identifying outliers. Box plots are commonly used to visualize the distribution and statistical summary of a numeric variable across different categories. They provide insights into the central tendency, spread, skewness, and potential outliers within each category, making them useful for comparing distributions and identifying differences between groups.
- sns.scatterplot -  catter plots are used to visualize the relationship between two continuous variables. They help identify patterns, clusters, or correlations between the variables.
- sns.lineplot - Line plots are useful for visualizing trends or changes over time. They are commonly used to display time series data or show the progression of a variable.

2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns 
import matplotlib.pyplot as plt
df = sns.load_dataset("fmri")
df.head()
print(df["event"].unique())
print(df["region"].unique())
```

```python
plt.title("FMRI signal by timepoint for different events and regions " )
sns.lineplot(data = df, x  = "timepoint" , y = "signal" , hue = "event" , style = "region")
```

3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
df2 = sns.load_dataset("titanic")
df2.head(2)
```

```python
plt.figure(figsize= (15,7))
plt.subplot(1,2,1)
sns.boxplot(data = df2 , x = "pclass" , y = "age")
plt.title("Box plot of Age by pclass ")

plt.subplot(1,2,2)
sns.boxplot(data = df2 , x = "pclass" , y = "fare")
plt.title("Box plot of fare by pclass ")
```

4. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset

```python
df3 = sns.load_dataset("diamonds")
df3.head(2)
```

```python
plt.figure(figsize= (15,7))
plt.title(" Diamond price with respect to cut " )
sns.histplot(data = df3 , x = "price" , hue = "cut" ,kde = True )
```
