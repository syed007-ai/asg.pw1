2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
df1 = pd.DataFrame({"A" : ["Virat" , "Maxwell" , "Faf"],
                   "B" : ["Dhoni" ,"Jadeja" , "Conway"],
                   "C" : ["Hardik" , "Shubhman" ,"Rashid"]})
```

```python
df1
```

```python
def reindex_dataframe (df) :
    new_index = pd.RangeIndex(start = 1 , stop = len(df)*2 , step = 2)
    df.set_index(new_index , inplace = True)
    return df 
```

```python
new_df = reindex_dataframe(df1)
new_df
```

calculate and print the sum of the first three values, which is 60.

5. How are DataFrame.size() and DataFrame.shape() different?

> DataFrame.size() and DataFrame.shape() are both methods in Pandas that provide information about the size or shape of a DataFrame, but they differ in what they represent.

* DataFrame.size() : returns the total number of elements in the DataFrame. It counts all the cells in the DataFrame, including both rows and columns. It gives you the total count of elements regardless of their values. For example, if you have a DataFrame with 3 rows and 4 columns, DataFrame.size() will return 12.
* DataFrame.shape() : returns a tuple , that represents the dimensions of the DataFrame. It tells you the number of rows and columns in the DataFrame. The first value in the tuple represents the number of rows, and the second value represents the number of columns. For example, if you have a DataFrame with 3 rows and 4 columns, DataFrame.shape() will return (3, 4).
>> eg :

```python
data = {'A': [1, 22, 30],
        'B': [4, 55, 60],
        'C': [7, 88, 90]}
df3 = pd.DataFrame(data)
df3
```

```python
print("The size of the DataFrame is :" ,df3.size) # returns the no.of elements
print("The shape of the DataFrame is :" ,df3.shape) # returns the no.of rows and col's as a tuple.
```

9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
df6 = pd.DataFrame({'Values': [44, 73, 92, 15, 20, 16, 8]})
df6
```

```python
def cal_stats(df) :
    mean_value = df6["Values"].mean()
    median_value = df6["Values"].median()
    std_value = df6["Values"].std()
    return mean_value , median_value, std_value
```

```python
mean, median , std = cal_stats(df6)
print("Mean : ", mean)
print("Median : " , median)
print("Std_dev : ", std )
```

Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
df7 = pd.DataFrame({'Sales': [10, 12, 15, 8, 10, 14, 13, 16, 18, 20, 12, 15],
                   'Date': pd.date_range(start='2023-01-01', periods=12)})
df7
```

```python
def moving_average(df) :
    df7["Moving Average"] = df7["Sales"].rolling(window = 7, min_periods= 1).mean()
    return df7
new_df4 = moving_average(df7)
new_df4
```

11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.

```python
df8 = pd.DataFrame({"Date" : pd.date_range(start = "2023-05-07" , periods = 12)})
df8
```

```python
def week_day(df) :
    df8["Week day"] = df8["Date"].dt.day_name()
    return df8
new_df5 = week_day(df8)
new_df5
```

12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
df9 = pd.DataFrame({"Date": pd.to_datetime(['2023-01-01 11:00:00', '2023-01-02 16:30:00', '2023-01-03 19:45:00', '2023-01-04 08:20:00' , "2023-02-02 06:30:00"]),
                   "Value" : [10,20,30,40,59]})
```

```python
df9
```

```python
def selecting_rows (df) :
    start_date = "2023-01-01"
    end_date = "2023-01-31"
    selected_rows = df9[(df9["Date"] >= start_date) & (df9["Date"] <= end_date)]
    return selected_rows
```

```python
select_rows = selecting_rows(df9)
select_rows
```
