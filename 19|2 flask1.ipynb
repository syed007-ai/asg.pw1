3. What is App routing in Flask? Why do we use app routes?


* When we use a web application, we send requests to a server (like clicking a link or filling out a form). In Flask, app routing is the way we handle those requests on the server side.

* Think of it like a map with different paths to different places. Each path (or URL) leads to a different page on the website. We use app routes in Flask to create those paths and tell the server what to do when someone visits each path.

* For eg, if someone visits the URL "/home" on a Flask web application, the server will know to display the homepage. If someone visits "/contact", the server will display the contact page.

* By using app routes, we can create a logical structure for our web application and make it easy to handle different requests from users. This helps us build more complex and functional websites.

5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

* In Flask, url_for() is a function that helps you generate URLs for different pages in your web application. You can use it to create links between different pages in your app, or to redirect the user to a different page.

* The url_for() function takes the name of a view function (which is a Python function that handles a specific web page or action) and returns the URL for that view function. It works by using the URL patterns you've defined in your Flask app, and generates a URL based on those patterns.
