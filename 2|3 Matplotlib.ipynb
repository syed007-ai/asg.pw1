2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3) #This means that each time you run the code, the random numbers generated by functions like np.random.normal will be the same, resulting in consistent data points for x and y. Setting a seed is often done to obtain reproducible results, especially when working with random processes and wanting consistent outcomes for testing or analysis purposes.
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y ,c="g")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("scatter plot")
```

>* A scatter plot is a type of plot that displays the relationship between two variables. It represents data points as individual dots on a two-dimensional plane, with each dot corresponding to a data point with its x and y values. 
> >  Scatter plots are useful for visualizing the correlation or pattern between two continuous variables.

3. Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
* import numpy as np
* For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
* For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
* For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
* For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

>>  subplot() function in Matplotlib is used to create multiple subplots within a single figure. It allows you to divide the figure into a grid of rows and columns and specify the position of each subplot. 

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1) #2 rows, 2col's , subplot 1
plt.plot(x1,y1)
plt.title("line 1")

plt.subplot(2,2,2) #2 rows, 2col's , subplot 2
plt.plot(x2,y2)
plt.title("line 2")

plt.subplot(2,2,3) #2 rows, 2col's , subplot 1
plt.plot(x3,y3)
plt.title("line 3")

plt.subplot(2,2,4) #2 rows, 2col's , subplot 1
plt.plot(x4,y4)
plt.title("line 4")

plt.tight_layout()
```

> Hence using matplotlib's subplot() we can visualise different plots in a single frame.

4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

* import numpy as np
* company = np.array(["Apple", "Microsoft", "Google", "AMD"])
* profit = np.array([3000, 8000, 1000, 10000])

> A bar graph is a kind of graph that represents categorical data with rectangular bars.
>>  Each bar represents a category, and the length of the bar corresponds to the value or frequency of that category. 
>* Bar plots are commonly used to compare and visualize the distribution or relationship between different categories or groups.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize= (4,4))
plt.bar(company ,profit ,color = "#FF5733")
plt.title(" Bar Plot ")
plt.xlabel("Name of the company ")
plt.ylabel(" profit ")
plt.tight_layout()
```

```python
plt.figure(figsize= (4,4))
plt.barh(company ,profit ,color = "#FF5733")
plt.title(" Horizontal Bar Plot ")
plt.xlabel(" profit ")
plt.ylabel(" Name of the company ")
plt.tight_layout()
```

>> In a bar plot /column plot/ vertical plot, the bars are plotted vertically along the x-axis, and their lengths represent the values being plotted. 
* The x-axis typically represents the categories or groups, while the y-axis represents the numerical values or frequencies associated with each category.
* Bar plots are commonly used to compare and visualize categorical data or to show the distribution of a variable across different categories.

> > In a horizontal bar plot, the bars are plotted horizontally along the y-axis, and their lengths still represent the values being plotted. 
* The y-axis represents the categories or groups, while the x-axis represents the numerical values or frequencies associated with each category. 
* Horizontal bar plots are useful when you have long category labels or want to emphasize the comparison of values across different categories by aligning the bars horizontally.



5. What is a box plot? Why is it used? Using the following data plot a box plot.
* box1 = np.random.normal(100, 10, 200)
* box2 = np.random.normal(90, 20, 200)

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
```

> A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays summary statistics such as the minimum, first quartile, median, third quartile, and maximum values, as well as any potential outliers.

> > Box plots are used to provide a visual summary of the data's distribution, including measures of central tendency and variability. They are particularly useful for comparing multiple datasets or visualizing the spread and skewness of a single dataset. The key components of a box plot are as follows:

* The box represents the interquartile range (IQR), which spans from the first quartile (Q1) to the third quartile (Q3). It contains the middle 50% of the data.
* The line within the box represents the median, which is the middle value of the dataset.
* The whiskers extend from the box to the minimum and maximum values of the dataset, excluding any outliers.
* Outliers, if present, are plotted as individual points outside the whiskers.
